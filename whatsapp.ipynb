```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up Chrome options (remove headless mode to see the browser)
options = Options()
# options.add_argument("--headless")  # Uncomment to run Chrome in headless mode (without opening the browser window)

# Set up ChromeDriver path
service = Service(executable_path=r"C:\webdriver\chromedriver.exe")  # Replace with your ChromeDriver path

# Initialize WebDriver
driver = webdriver.Chrome(service=service, options=options)

# Open WhatsApp Web
driver.get("https://web.whatsapp.com/")

# Wait for the QR code to be scanned or for the page to load
print("Scan the QR code and wait for the page to load...")
try:
    # Wait for WhatsApp Web to load and the chat icon to appear
    WebDriverWait(driver, 90).until(EC.presence_of_element_located((By.XPATH, '//span[@data-icon="chat"]')))
    print("WhatsApp Web is working!")
except Exception as e:
    print("Failed to load WhatsApp Web:", e)
    driver.quit()  # Exit the script if the page doesn't load

# Read contacts from Excel file
file_path = "C:\\Users\\Uzair Khan\\Downloads\\aaaa.xlsx"  # Replace with your actual file path
df = pd.read_excel(file_path)

# Attachments list (use actual paths to your files)
attachments = [
    r"C:\Users\Uzair Khan\Downloads\DOC-20250823-WA0034_250823_155926.pdf",  # Replace with actual file paths
    r"C:\Users\Uzair Khan\Downloads\Kiran Temkar.pdf",
    r"C:\Users\Uzair Khan\Downloads\WhatsApp Image 2025-08-23 at 9.09.48 PM.jpeg",
    r"C:\Users\Uzair Khan\Downloads\WhatsApp Image 2025-08-23 at 9.09.49 PM.jpeg"
]

# Message to send
message = "Hello, this is a test message with 4 attachments."

# Retry mechanism to open chat
def open_chat(contact):
    retries = 3  # Retry 3 times
    for attempt in range(retries):
        try:
            search_box = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, '//div[@contenteditable="true"][@data-tab="3"]')))
            search_box.send_keys(contact)
            time.sleep(2)
            search_box.send_keys(Keys.RETURN)
            time.sleep(2)
            return True
        except Exception as e:
            print(f"Attempt {attempt + 1} to open chat with {contact} failed: {e}")
            if attempt == retries - 1:
                print(f"Failed to open chat with {contact}. Skipping.")
            time.sleep(5)
    return False

# Loop through contacts and send messages with attachments
for index, row in df.iterrows():
    contact = row['Phone Number']  # Ensure the column name matches exactly

    # Open the chat with the contact
    if not open_chat(contact):
        continue  # Skip this contact and move to the next one

    # Send attachments one by one
    for file_path in attachments:
        try:
            # Wait for the attachment button (clip icon) to be clickable
            attach_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//div[@title="Attach"]')))
            attach_button.click()
            
            # Wait for the file input element to be available and send the file path
            file_input = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, '//input[@type="file"]')))
            file_input.send_keys(file_path)
            time.sleep(2)  # Wait for file to upload

            # Wait for the send button to be clickable and click it
            send_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//span[@data-icon="send"]')))
            send_button.click()
            print(f"Attachment {file_path} sent to {contact}")
            time.sleep(2)  # Add delay before sending the next file
        except Exception as e:
            print(f"Error while sending attachment {file_path} to {contact}: {e}")

    # Send the text message after attachments
    try:
        message_box = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, '//div[@contenteditable="true"][@data-tab="1"]')))
        message_box.send_keys(message)
        message_box.send_keys(Keys.RETURN)
        print(f"Message sent to {contact}")
    except Exception as e:
        print(f"Error while sending message to {contact}: {e}")

    time.sleep(3)  # Wait before moving to the next contact

# Close the browser after sending all messages
driver.quit()
print("All messages with attachments have been sent!")
```
